```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,accuracy_score
```

##Predicting the Rate of Accidents by using company provided data with 22 different sensors recognized patterns

```python
train = pd.read_csv("https://raw.githubusercontent.com/Prasad-Mamidi/Machine_Learning/main/Datasets/Car_Accident/car_train.csv")
test = pd.read_csv("https://raw.githubusercontent.com/Prasad-Mamidi/Machine_Learning/main/Datasets/Car_Accident/car_test.csv")
```

```python
train.head()
```

```python
train.describe()
```

```python
train.info()
```

```python
X = train.drop("Fatal",axis=1)
y = train["Fatal"]
dct = DecisionTreeClassifier(max_depth=6)
dct = dct.fit(X,y)
print(classification_report(y,dct.predict(X)))
```

```python
print(classification_report(test["Fatal"],dct.predict(test.drop("Fatal",axis=1))))
print(dct.score(test.drop("Fatal",axis=1),test["Fatal"]))
```

```python
rf = RandomForestClassifier(n_estimators=50,max_features=4,max_depth=9)
rf = rf.fit(X,y)
print(accuracy_score(y,rf.predict(X)))
print(accuracy_score(test["Fatal"],rf.predict(test.drop("Fatal",axis=1))))
```

```python
rf = RandomForestClassifier(n_estimators=50,max_features=6,max_depth=10)
rf = rf.fit(X,y)
print(accuracy_score(y,rf.predict(X)))
print(accuracy_score(test["Fatal"],rf.predict(test.drop("Fatal",axis=1))))
```

```python
print(f"Train Accuracy : {accuracy_score(y,rf.predict(X))}")
print(f"Test Accuracy  : {accuracy_score(test['Fatal'], rf.predict(test.drop('Fatal',axis=1)))}")
```
